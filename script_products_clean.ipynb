```python
import pandas as pd
```

```python
# chargement du csv dans un dataframe
df_products = pd.read_csv("castorama_products.csv")
```

```python
# suppression des espaces en trop par colonne
df_products['unique_id'] = df_products['unique_id'].str.strip()
df_products['category'] = df_products['category'].str.strip()
df_products['subcategory'] = df_products['subcategory'].str.strip()
df_products['subsubcategory'] = df_products['subsubcategory'].str.strip()
df_products['title'] = df_products['title'].str.strip()
df_products['price'] = df_products['price'].str.strip()
df_products['url'] = df_products['url'].str.strip()
```

```python
# suppression des accents sur les voyelles de la colonne "unique_id"
df_products['unique_id'] = df_products['unique_id'].str.replace("à", "a")
df_products['unique_id'] = df_products['unique_id'].str.replace("ë", "e")
df_products['unique_id'] = df_products['unique_id'].str.replace("é", "e")
df_products['unique_id'] = df_products['unique_id'].str.replace("è", "e")
df_products['unique_id'] = df_products['unique_id'].str.replace("ê", "e")
df_products['unique_id'] = df_products['unique_id'].str.replace("ï", "i")
df_products['unique_id'] = df_products['unique_id'].str.replace("î", "i")
df_products['unique_id'] = df_products['unique_id'].str.replace("ô", "o")
df_products['unique_id'] = df_products['unique_id'].str.replace("ù", "u")
```

```python
# vérification des doublons
df_products.duplicated(subset=["unique_id"]).sum()
```

```python
# vérification des valeurs manquantes
df_products['unique_id'].isna().sum()
```

```python
# vérification des types des colonnes
df_products.dtypes
```

```python
# remplacer les virgules de la colonne "price" par des points
df_products["price"] = df_products["price"].str.replace(",", ".")
```

```python
# vérifier les valeurs non-convertible en float
df_products[df_products["price"] == "(None.)"]
```

```python
# suppression des lignes dont la valeur dans la colonne 'price' n'est pas un float 
df_products_filter = df_products.loc[df_products["price"] != "(None.)"]
df_products_filter
```

```python
#suppression des espaces vides
df_products_filter["price"] = df_products_filter["price"].str.replace(" ", "")
df_products_filter["price"] = df_products_filter["price"].astype(float)
```

```python
#Conversion vers un csv
df_products_filter.to_csv("castorama_products_clean.csv", index=False)
```
